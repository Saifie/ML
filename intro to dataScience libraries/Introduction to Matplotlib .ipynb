```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease["age"].plot.hist()
```

```python
heart_disease.head()["age"].plot(figsize=(10,5),x="chol",y="age",title="heart");
```

```python
heart_disease.head().plot(figsize=(20,50),kind="hist",subplots=True);
```

```python
over_50=heart_disease[heart_disease["age"] > 50]
```

```python
over_50
```

```python
fig,ax=plt.subplots(figsize=(15,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
```

```python
fig,ax=plt.subplots(figsize=(15,6))
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax.set(title="Heart diease over 50",xlabel="Age",ylabel="Cholestrol");
ax.legend(*scatter.legend_elements(),title="target")
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1,figsize=(15,10))

scatter1=ax1.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")
scatter2=ax2.scatter(x=over_50["age"],y=over_50["thalach"],cmap="winter",c=over_50["target"])
ax1.set(title="Heart diease over 50",xlabel="Age",ylabel="Cholestrol");
ax2.set(title="Heart diease over 50",xlabel="Age",ylabel="thalach");
ax1.legend(*scatter.legend_elements(),title="target")
ax2.legend(*scatter.legend_elements(),title="target")

ax1.axhline(over_50["chol"].mean(),linestyle="--");
ax2.axhline(over_50["thalach"].mean(),linestyle="--");


```
